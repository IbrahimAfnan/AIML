```python
import pandas as pd
d=pd.read_csv('/content/day.csv')
print(d)
```

```python
x1=d['season']
y=d['count']
print(x1,y)
```

```python
ee=[]
import numpy as np
mm=[]
cc=[]
def liner(m1,c):
  ye=[]
  yp=[]
  sum=0
  mm.append(m1)
  cc.append(c)
  for i in range(0,len(x1)):
    yp.append(m1*x1[i]+c)
    ye.append((y[i]-yp[i])**2)
  for i in range(0,len(x1)):
    sum=sum+ye[i]
  ee1=np.mod(sum,len(x1))
  ee.append(ee1)
  print(ee1)
```

```python
a=np.array(ee).min()
print(a)
ind=ee.index(a)
print(ind)
print(mm[ind],cc[ind])
print(len(ee))
print("lll",ee[42])
```

```python
for i in range(0,50):
  liner(i+10,i+20)
```

```python
from matplotlib import pyplot as pt
import numpy as np
x=[]
for i in range(0,50):
  x.append(i)
pt.plot(x,ee)
```
